```python
def takeMean(dataSet):
    total = 0
    num = len(dataSet)
    
    for i in dataSet:
        total += float(i)
    return total / num

def takeStd(dataSet, mean):
    sumOf = 0
    num = len(dataSet)
    
    for i in dataSet:
        sumOf += (i - mean) ** 2
    return np.sqrt(sumOf / (num - 1))

meanTempOut = takeMean(tempDataOut["temp"])
stdTempOut = takeStd(tempDataOut["temp"], takeMean(tempDataOut["temp"]))
uncTempOut =  stdTempOut / np.sqrt(len(tempDataOut["temp"]))
meanTempIn = takeMean(tempDataIn["temp"])
stdTempIn = takeStd(tempDataIn["temp"], takeMean(tempDataIn["temp"]))
uncTempIn = stdTempIn / np.sqrt(len(tempDataIn["temp"]))
stdErrTemp = np.sqrt((stdTempOut**2 / len(tempDataOut["temp"])) + (stdTempIn**2 / len(tempDataIn["temp"])))
print(abs(meanTempOut - meanTempIn) > (3 * stdErrTemp))

meanGasOut = takeMean(gasDataOut["gas"])
stdGasOut = takeStd(gasDataOut["gas"], takeMean(gasDataOut["gas"]))
uncGasOut = stdGasOut / np.sqrt(len(gasDataOut["gas"]))
meanGasIn = takeMean(gasDataIn["gas"])
stdGasIn = takeStd(gasDataIn["gas"], takeMean(gasDataIn["gas"]))
uncGasIn = stdGasIn / np.sqrt(len(gasDataIn["gas"]))
stdErrGas = np.sqrt((stdGasOut**2 / len(gasDataOut["gas"])) + (stdGasIn**2 / len(gasDataIn["gas"])))
print(abs(meanGasOut - meanGasIn) > (3 * stdErrGas))

meanHumOut = takeMean(humDataOut["humidity"])
stdHumOut = takeStd(humDataOut["humidity"], takeMean(humDataOut["humidity"]))
uncHumOut = stdHumOut / np.sqrt(len(humDataOut["humidity"]))
meanHumIn = takeMean(humDataIn["humidity"])
stdHumIn = takeStd(humDataIn["humidity"], takeMean(humDataIn["humidity"]))
uncHumIn = stdHumIn / np.sqrt(len(humDataIn["humidity"]))
stdErrHum = np.sqrt((stdHumOut**2 / len(humDataOut["humidity"])) + (stdHumIn**2 / len(humDataIn["humidity"])))
print(abs(meanHumOut - meanHumIn) > (3 * stdErrHum))

meanPresOut = takeMean(presDataOut["pressure"])
stdPresOut = takeStd(presDataOut["pressure"], takeMean(presDataOut["pressure"]))
uncPresOut = stdPresOut / np.sqrt(len(presDataOut["pressure"]))
meanPresIn = takeMean(presDataIn["pressure"])
stdPresIn = takeStd(presDataIn["pressure"], takeMean(presDataIn["pressure"]))
uncPresIn = stdPresOut / np.sqrt(len(presDataIn["pressure"]))
stdErrPres = np.sqrt((stdPresOut**2 / len(presDataOut["pressure"])) + (stdPresIn**2 / len(presDataIn["pressure"])))
print(abs(meanPresOut - meanPresIn) > (3 * stdErrPres))

meanAltOut = takeMean(altDataOut["alt"])
stdAltOut = takeStd(altDataOut["alt"], takeMean(altDataOut["alt"]))
uncAltOut = stdAltOut / np.sqrt(len(altDataOut["alt"]))
meanAltIn = takeMean(altDataIn["alt"])
stdAltIn = takeStd(altDataIn["alt"], takeMean(altDataIn["alt"]))
uncAltIn = stdAltIn / np.sqrt(len(altDataIn["alt"]))
stdErrAlt = np.sqrt((stdAltOut**2 / len(altDataOut["alt"])) + (stdAltIn**2 / len(altDataIn["alt"])))
print(abs(meanAltOut - meanAltIn) > (3 * stdErrAlt))

meanStd10Out = takeMean(std10DataOut["stndr10"])
stdStd10Out = takeStd(std10DataOut["stndr10"], takeMean(std10DataOut["stndr10"]))
uncStd10Out = stdStd10Out / np.sqrt(len(std10DataOut["stndr10"]))
meanStd10In = takeMean(std10DataIn["stndr10"])
stdStd10In = takeStd(std10DataIn["stndr10"], takeMean(std10DataIn["stndr10"]))
uncStd10In = stdStd10Out / np.sqrt(len(std10DataIn["stndr10"]))
stdErrStd10 = np.sqrt((stdStd10Out**2 / len(std10DataOut["stndr10"])) + (stdStd10In**2 / len(std10DataIn["stndr10"])))
print(abs(meanStd10Out - meanStd10In) > (3 * stdErrStd10))

meanStd25Out = takeMean(std25DataOut["stndr25"])
stdStd25Out = takeStd(std25DataOut["stndr25"], takeMean(std25DataOut["stndr25"]))
uncStd25Out = stdStd25Out / np.sqrt(len(std25DataOut["stndr25"]))
meanStd25In = takeMean(std25DataIn["stndr25"])
stdStd25In = takeStd(std25DataIn["stndr25"], takeMean(std25DataIn["stndr25"]))
uncStd25In = stdStd25In / np.sqrt(len(std25DataIn["stndr25"]))
stdErrStd25 = np.sqrt((stdStd25Out**2 / len(std25DataOut["stndr25"])) + (stdStd25In**2 / len(std25DataIn["stndr25"])))
print(abs(meanStd25Out - meanStd25In) > (3 * stdErrStd25))

meanStd100Out = takeMean(std100DataOut["stndr100"])
stdStd100Out = takeStd(std100DataOut["stndr100"], takeMean(std100DataOut["stndr100"]))
uncStd100Out = stdStd100Out / np.sqrt(len(std100DataOut["stndr100"]))
meanStd100In = takeMean(std100DataIn["stndr100"])
stdStd100In = takeStd(std100DataIn["stndr100"], takeMean(std100DataIn["stndr100"]))
uncStd100In = stdStd100In / np.sqrt(len(std100DataIn["stndr100"]))
stdErrStd100 = np.sqrt((stdStd100Out**2 / len(std100DataOut["stndr100"])) + (stdStd100In**2 / len(std100DataIn["stndr100"])))
print(abs(meanStd100Out - meanStd100In) > (3 * stdErrStd100))

meanEnv10Out = takeMean(env10DataOut["enviro10"])
stdEnv10Out = takeStd(env10DataOut["enviro10"], takeMean(env10DataOut["enviro10"]))
uncEnv10Out = stdEnv10Out / np.sqrt(len(env10DataOut["enviro10"]))
meanEnv10In = takeMean(env10DataIn["enviro10"])
stdEnv10In = takeStd(env10DataIn["enviro10"], takeMean(env10DataIn["enviro10"]))
uncEnv10In = stdEnv10In / np.sqrt(len(env10DataIn["enviro10"]))
stdErrEnv10 = np.sqrt((stdEnv10Out**2 / len(env10DataOut["enviro10"])) + (stdEnv10In**2 / len(env10DataIn["enviro10"])))
print(abs(meanEnv10Out - meanEnv10In) > (3 * stdErrEnv10))

meanEnv25Out = takeMean(env25DataOut["enviro25"])
stdEnv25Out = takeStd(env25DataOut["enviro25"], takeMean(env25DataOut["enviro25"]))
uncEnv25Out = stdEnv25Out / np.sqrt(len(env25DataOut["enviro25"]))
meanEnv25In = takeMean(env25DataIn["enviro25"])
stdEnv25In = takeStd(env25DataIn["enviro25"], takeMean(env25DataIn["enviro25"]))
uncEnv25In = stdEnv25In / np.sqrt(len(env25DataIn["enviro25"]))
stdErrEnv25 = np.sqrt((stdEnv25Out**2 / len(env25DataOut["enviro25"])) + (stdEnv25In**2 / len(env25DataIn["enviro25"])))
print(abs(meanEnv25Out - meanEnv25In) > (3 * stdErrEnv25))

meanEnv100Out = takeMean(env100DataOut["enviro100"])
stdEnv100Out = takeStd(env100DataOut["enviro100"], takeMean(env100DataOut["enviro100"]))
uncEnv100Out = stdEnv100Out / np.sqrt(len(env100DataOut["enviro100"]))
meanEnv100In = takeMean(env100DataIn["enviro100"])
stdEnv100In = takeStd(env100DataIn["enviro100"], takeMean(env100DataIn["enviro100"]))
uncEnv100In = stdEnv100In / np.sqrt(len(env100DataIn["enviro100"]))
stdErrEnv100 = np.sqrt((stdEnv100Out**2 / len(env100DataOut["enviro100"])) + (stdEnv100In**2 / len(env100DataIn["enviro100"])))
print(abs(meanEnv100Out - meanEnv100In) > (3 * stdErrEnv100))

meansP03Out = takeMean(p03DataOut["p03um"])
stdP03Out = takeStd(p03DataOut["p03um"], takeMean(p03DataOut["p03um"]))
uncP03Out = stdP03Out / np.sqrt(len(p03DataOut["p03um"]))
meansP03In = takeMean(p03DataIn["p03um"])
stdP03In = takeStd(p03DataIn["p03um"], takeMean(p03DataIn["p03um"]))
uncP03In = stdP03In / np.sqrt(len(p03DataIn["p03um"]))
stdErrP03 = np.sqrt((stdP03Out**2 / len(p03DataOut["p03um"])) + (stdP03In**2 / len(p03DataIn["p03um"])))
print(abs(meansP03Out - meansP03In) > (3 * stdErrP03))

meansP05Out = takeMean(p05DataOut["p05um"])
stdP05Out = takeStd(p05DataOut["p05um"], takeMean(p05DataOut["p05um"]))
uncP05Out = stdP05Out / np.sqrt(len(p05DataOut["p05um"]))
meansP05In = takeMean(p05DataIn["p05um"])
stdP05In = takeStd(p05DataIn["p05um"], takeMean(p05DataIn["p05um"]))
uncP05In = stdP05In / np.sqrt(len(p05DataIn["p05um"]))
stdErrP05 = np.sqrt((stdP05Out**2 / len(p05DataOut["p05um"])) + (stdP05In**2 / len(p05DataIn["p05um"])))
print(abs(meansP05Out - meansP05In) > (3 * stdErrP05))

meansP10Out = takeMean(p10DataOut["p10um"])
stdP10Out = takeStd(p10DataOut["p10um"], takeMean(p10DataOut["p10um"]))
uncP10Out = stdP10Out / np.sqrt(len(p10DataOut["p10um"]))
meansP10In = takeMean(p10DataIn["p10um"])
stdP10In = takeStd(p10DataIn["p10um"], takeMean(p10DataIn["p10um"]))
uncP10In = stdP10In / np.sqrt(len(p10DataIn["p10um"]))
stdErrP10 = np.sqrt((stdP10Out**2 / len(p10DataOut["p10um"])) + (stdP10In**2 / len(p10DataIn["p10um"])))
print(abs(meansP10Out - meansP10In) > (3 * stdErrP10))

meansP25Out = takeMean(p25DataOut["p25um"])
stdP25Out = takeStd(p25DataOut["p25um"], takeMean(p25DataOut["p25um"]))
uncP25Out = stdP25Out / np.sqrt(len(p25DataOut["p25um"]))
meansP25In = takeMean(p25DataIn["p25um"])
stdP25In = takeStd(p25DataIn["p25um"], takeMean(p25DataIn["p25um"]))
uncP25In = stdP25In / np.sqrt(len(p25DataIn["p25um"]))
stdErrP25 = np.sqrt((stdP25Out**2 / len(p25DataOut["p25um"])) + (stdP25In**2 / len(p25DataIn["p25um"])))
print(abs(meansP25Out - meansP25In) > (3 * stdErrP25))

meansP50Out = takeMean(p50DataOut["p50um"])
stdP50Out = takeStd(p50DataOut["p50um"], takeMean(p50DataOut["p50um"]))
uncP50Out = stdP50Out / np.sqrt(len(p50DataOut["p50um"]))
meansP50In = takeMean(p50DataIn["p50um"])
stdP50In = takeStd(p50DataIn["p50um"], takeMean(p50DataIn["p50um"]))
uncP50In = stdP50In / np.sqrt(len(p50DataIn["p50um"]))
stdErrP50 = np.sqrt((stdP50Out**2 / len(p50DataOut["p50um"])) + (stdP50In**2 / len(p50DataIn["p50um"])))
print(abs(meansP50Out - meansP50In) > (3 * stdErrP50))

meansP100Out = takeMean(p100DataOut["p100um"])
stdP100Out = takeStd(p100DataOut["p100um"], takeMean(p100DataOut["p100um"]))
uncP100Out = stdP100Out / np.sqrt(len(p100DataOut["p100um"]))
meansP100In = takeMean(p100DataIn["p100um"])
stdP100In = takeStd(p100DataIn["p100um"], takeMean(p100DataIn["p100um"]))
uncP100In = stdP100In / np.sqrt(len(p100DataIn["p100um"]))
stdErrP100 = np.sqrt((stdP100Out**2 / len(p100DataOut["p100um"])) + (stdP100In**2 / len(p100DataIn["p100um"])))
print(abs(meansP100Out - meansP100In) > (3 * stdErrP100))
```
