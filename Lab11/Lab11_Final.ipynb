```python
#Thorium
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

thoriumData = []
thor0 = pd.read_csv("Lab11ThRCorrected.csv")
thoriumData.append(thor0)
thor1 = pd.read_csv("Lab11ThW1Corrected.csv")
thoriumData.append(thor1)
thor2 = pd.read_csv("Lab11ThW2Corrected.csv")
thoriumData.append(thor2)
thor3 = pd.read_csv("Lab11ThW3Corrected.csv")
thoriumData.append(thor3)
thor4 = pd.read_csv("Lab11BgCorrected.csv")
thoriumData.append(thor4)

plt.figure(figsize = (10, 6))

pos = 0
labelList = ["No", "1", "2", "3"]
colors = ["b", "orange", "g", "r"]
for i in thoriumData[0:4]: #iterate each dataset
    plt.hist(i.iloc[:,1], alpha = 0.5, bins = np.linspace(0, 600, 30), label = labelList[pos] + " W")
    plt.axvline(i.iloc[:,1].mean(), color = colors[pos], linestyle='--', label = "Mean " + labelList[pos] + " W")
    print(labelList[pos] + " Tungsten:   Mean = " + str(round(i.iloc[:,1].mean(), 2)) + ", Std Dev = " + str(round(i.iloc[:,1].std(), 2)))
    pos += 1

plt.title("CPS vs Frequency")
plt.xlabel("Counts")
plt.xlim(0, 175)
plt.ylabel("Frequency of Counts")
plt.legend()
plt.show()
```

### How do the mean (μ) and standard deviation (σ) compare?
##### The mean varies drastically compared to the two sources, with the means of each source decreasing, but Tungsten's mean decreases more per plate added than Aluminium. The standard deviation for Aluminium remains at an approximate and constant 30% of the mean, while Tungsten's standard deviation varies from ~34% to ~16% of the respective means.
#### a. How do the measured standard deviations compare to the theoretical uncertainty you would expect from counting statistics?
##### The measured standard deviations for the lower means (29 and 32.38) align closely with theoretical counting statistics. In contrast, the measurement for the highest mean (101.54) shows significantly higher variation than theoretically expected based on the square root of the mean.
#### b. What general trend do you see in the counts measured as a function of the thickness of absorber material?
##### I see the counts decrease almost exponentially, as the count rate becomes more than halved, then a quarter removed, etc., for Tungsten and very similarly for Aluminium, except for the radon increase in counts. But typically, increased thickness means fewer counts measured.

```python
#Fiestaware
bgDataRaw = pd.read_csv('Lab11BgCorrected.csv')
fiestaRaw = pd.read_csv('Lab11FiestaRCorrected.csv')
fiestaAl1Raw = pd.read_csv('Lab11FiestaAlCorrected.csv')
fiestaAl2Raw = pd.read_csv('Lab11FiestaAl2Corrected.csv')
fiestaAl3Raw = pd.read_csv('Lab11FiestaAl3Corrected.csv')

bgData = pd.DataFrame()
bgData['elapsedTime'] = bgDataRaw['elapsedTime']
bgData['countPerInterval'] = bgDataRaw['totalCount'].diff().fillna(0)

fiestaRData = pd.DataFrame()
fiestaRData['elapsedTime'] = fiestaRaw['elapsedTime']
fiestaRData['countPerInterval'] = fiestaRaw['totalCount'].diff().fillna(0)

fiestaAl1Data = pd.DataFrame()
fiestaAl1Data['elapsedTime'] = fiestaAl1Raw['elapsedTime']
fiestaAl1Data['countPerInterval'] = fiestaAl1Raw['totalCount'].diff().fillna(0)

fiestaAl2Data = pd.DataFrame()
fiestaAl2Data['elapsedTime'] = fiestaAl2Raw['elapsedTime']
fiestaAl2Data['countPerInterval'] = fiestaAl2Raw['totalCount'].diff().fillna(0)

fiestaAl3Data = pd.DataFrame()
fiestaAl3Data['elapsedTime'] = fiestaAl3Raw['elapsedTime']
fiestaAl3Data['countPerInterval'] = fiestaAl3Raw['totalCount'].diff().fillna(0)

fiestaData = pd.DataFrame()
fiestaData['elapsedTime'] = fiestaRaw['elapsedTime']
fiestaData['rawCPI'] = fiestaRData['countPerInterval']
fiestaData['Al1CPI'] = fiestaAl1Data['countPerInterval']
fiestaData['Al2CPI'] = fiestaAl2Data['countPerInterval']
fiestaData['Al3CPI'] = fiestaAl3Data['countPerInterval']

plt.figure(figsize = (10, 6))
plt.xlabel('Counts')
plt.ylabel('Frequency of Counts')
plt.xlim(200,600)
bins = np.linspace(0, 600, 30)

data0 = fiestaData['rawCPI']
mu0, std0 = data0.mean(), data0.std()
plt.hist(fiestaData['rawCPI'], alpha = 0.5, bins = bins, label = 'No Al')
plt.axvline(mu0, linestyle='--', label = 'Mean No Al')

data1 = fiestaData['Al1CPI']
mu1, std1 = data1.mean(), data1.std()
plt.hist(fiestaData['Al1CPI'], alpha = 0.5, bins = bins, label = '1 Al')
plt.axvline(mu1, color = 'orange', linestyle='--', label = 'Mean 1 Al')

data2 = fiestaData['Al2CPI']
mu2, std2 = data2.mean(), data2.std()
plt.hist(fiestaData['Al2CPI'], alpha = 0.5, bins = bins, label = '2 Al')
plt.axvline(mu2, color = 'green', linestyle='--', label = 'Mean 2 Al')

data3 = fiestaData['Al3CPI']
mu3, std3 = data3.mean(), data3.std()
plt.hist(fiestaData['Al3CPI'], alpha = 0.5, bins = bins, label = '3 Al')
plt.axvline(mu3, color = 'red', linestyle='--', label = 'Mean 3 Al')

plt.legend()
plt.show()

print(f"No Aluminum:   Mean = {fiestaData['rawCPI'].mean():.2f}, Std Dev = {fiestaData['rawCPI'].std():.2f}")
print(f"1 Al Sheet:    Mean = {fiestaData['Al1CPI'].mean():.2f}, Std Dev = {fiestaData['Al1CPI'].std():.2f}")
print(f"2 Al Sheets:   Mean = {fiestaData['Al2CPI'].mean():.2f}, Std Dev = {fiestaData['Al2CPI'].std():.2f}")
print(f"3 Al Sheets:   Mean = {fiestaData['Al3CPI'].mean():.2f}, Std Dev = {fiestaData['Al3CPI'].std():.2f}")
```

```python
plt.figure(figsize = (10, 6))
thickness = [0, 1, 2, 3] # for # sheets
Almean = [fiestaData['rawCPI'].mean(), fiestaData['Al1CPI'].mean(), fiestaData['Al2CPI'].mean(), fiestaData['Al3CPI'].mean()]
Alstd = [fiestaData['rawCPI'].std(), fiestaData['Al1CPI'].std(), fiestaData['Al2CPI'].std(), fiestaData['Al3CPI'].std()]

Wmean = []
Wstd = []
for i in thoriumData[0:4]:
    Wmean.append(i.iloc[:,1].mean())
    Wstd.append(i.iloc[:,1].std())

plt.errorbar(thickness, Wmean, yerr = Wstd, fmt = 'o-', capsize = 5, label = 'Error (W)', color = 'pink', markersize = 8)
plt.errorbar(thickness, Almean, yerr = Alstd, fmt = 'o-', capsize = 5, label = 'Error (Al)', color = 'blue', markersize = 8)
plt.yscale('log')
plt.xlabel('Absorber Thickness # Sheets')
plt.ylabel('Mean Counts (Log)')
plt.legend()
plt.show()
```
