```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

thoriumData = []
thor0 = pd.read_csv("Lab11ThR.csv")
thoriumData.append(thor0)
thor1 = pd.read_csv("Lab11ThW1.csv")
thoriumData.append(thor1)
thor2 = pd.read_csv("Lab11ThW2.csv")
thoriumData.append(thor2)
thor3 = pd.read_csv("Lab11ThW3.csv")
thoriumData.append(thor3)
thor4 = pd.read_csv("Lab11Bg.csv")
thoriumData.append(thor4)

plt.figure(figsize = (10, 6))

pos = 0
labelList = ["No", "1", "2", "3"]
colors = ["b", "orange", "g", "r"]
for i in thoriumData[0:4]: #iterate each dataset
    plt.hist(i.iloc[:,1], alpha = 0.5, bins = np.linspace(0, 600, 30), label = labelList[pos] + " W")
    plt.axvline(i.iloc[:,1].mean(), color = colors[pos], linestyle='--', label = "Mean " + labelList[pos] + " W")
    print(labelList[pos] + " Tungsten:   Mean = " + str(round(i.iloc[:,1].mean(), 2)) + ", Std Dev = " + str(round(i.iloc[:,1].std(), 2)))
    pos += 1

plt.title("CPS vs Frequency")
plt.xlabel("Counts")
plt.xlim(0, 175)
plt.ylabel("Frequency of Counts")
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10, 6))

pos = 0
labelList = ["No", "1", "2", "3"]
colors = ["b", "orange", "g", "r"]
for i in thoriumData[0:4]:
    plt.errorbar(pos, i.iloc[:,1].mean(), yerr = i.iloc[:,1].std(), fmt = "o-", capsize = 5, label = labelList[pos] + " W Error", color = colors[pos], markersize = 8)
    pos += 1

plt.yscale('log')
plt.xlabel('Absorber Thickness # Al Sheets')
plt.ylabel('Mean Counts (Log)')
plt.legend()
plt.show()
```
