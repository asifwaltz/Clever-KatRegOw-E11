```python
import csv #comma seperated var
import time
import numpy as np

file = open("test.csv", "w", newline = None)
csvwriter = csv.writer(file, delimiter = ",")

meta = ["time", "data"]
csvwriter.writerow(meta)

for i in range(10):
    now = time.time()
    val = np.random.random()
    csvwriter.writerow([now, val])

file.close()
```
