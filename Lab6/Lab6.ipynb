```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = []
dataOurs = []
dataOutsideOurs = pd.read_csv("Lab6/Lab6OursOut.csv")
dataOurs.append(dataOutsideOurs)
dataInsideOurs = pd.read_csv("Lab6/Lab6OursIn.csv")
dataOurs.append(dataInsideOurs)
for i in range(2):
    dataOurs[i] = dataOurs[i].iloc[145 : 441, :]
    data.append(dataOurs[i])
    dataOurs[i] = dataOurs[i].drop(columns = ["stndr10", "stndr25", "stndr100", "enviro10", "enviro25", "enviro100"])

dataTheirs = []
dataOutsideTheirs = pd.read_csv("Lab6/Lab6TheirsOut.csv")
dataTheirs.append(dataOutsideTheirs)
dataInsideTheirs = pd.read_csv("Lab6/Lab6TheirsIn.csv")
dataTheirs.append(dataInsideTheirs)

labelList = ["Time", "Temperature", "Gas", "Humidity", "Pressure", "Altitude", "03 um Particles", "05 um Particles", "10 um Particles", "25 um Particles", "50 um Particles", "100 um Particles"]

dataTheirsList = []
dataTheirsList0 = list(dataTheirs[0])
dataTheirsList.append(dataTheirsList0)
dataTheirsList1 = list(dataTheirs[1])
dataTheirsList.append(dataTheirsList1)
for i in range(2):
    dataTheirsList[i][1], dataTheirsList[i][7] = dataTheirsList[i][7], dataTheirsList[i][1]
    dataTheirsList[i][2], dataTheirsList[i][8] = dataTheirsList[i][8], dataTheirsList[i][2]
    dataTheirsList[i][3], dataTheirsList[i][9] = dataTheirsList[i][9], dataTheirsList[i][3]
    dataTheirsList[i][4], dataTheirsList[i][10] = dataTheirsList[i][10], dataTheirsList[i][4]
    dataTheirsList[i][5], dataTheirsList[i][11] = dataTheirsList[i][11], dataTheirsList[i][5]
    dataTheirsList[i][6], dataTheirsList[i][7] = dataTheirsList[i][7], dataTheirsList[i][6]
    dataTheirsList[i][7], dataTheirsList[i][8] = dataTheirsList[i][8], dataTheirsList[i][7]
    dataTheirsList[i][8], dataTheirsList[i][9] = dataTheirsList[i][9], dataTheirsList[i][8]
    dataTheirsList[i][9], dataTheirsList[i][10] = dataTheirsList[i][10], dataTheirsList[i][9]
    dataTheirsList[i][10], dataTheirsList[i][11] = dataTheirsList[i][11], dataTheirsList[i][10]
    dataTheirs[i].columns = dataTheirsList[i]

#need to align the time in order for it to appear correctly
for i in range(1, dataOurs[0].shape[1]):
    plt.plot(dataOurs[0].iloc[:, 0], dataOurs[0].iloc[:, i], label = labelList[i] + " Outside Ours")
    plt.plot(dataTheirs[0].iloc[1:297, 0], dataTheirs[0].iloc[1:297, i], label = labelList[i] + " Outside Theirs")
    plt.xlabel("Time")
    plt.xticks([0, 295])
    plt.ylabel(labelList[i])
    plt.title(labelList[i] + " vs Time Outside")
    plt.legend()
    plt.show()
    
    plt.plot(dataOurs[1].iloc[:, 0], dataOurs[1].iloc[:, i], label = labelList[i] + " Inside Ours")
    plt.plot(dataTheirs[1].iloc[1:297, 0], dataTheirs[1].iloc[1:297, i], label = labelList[i] + " Inside Theirs")
    plt.xlabel("Time")
    plt.xticks([0, 295])
    plt.ylabel(labelList[i])
    plt.title(labelList[i] + " vs Time Inside")
    plt.legend()
    plt.show()
```
